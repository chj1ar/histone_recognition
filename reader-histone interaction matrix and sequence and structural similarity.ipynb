## Reader domain sequence similarity (problematic)

```python
readers = []
import os
for reader_filename in os.listdir('../data/reader_sequences/'):
    with open('../data/reader_sequences/{}'.format(reader_filename), 'r') as reader_file:
        reader = ''.join(reader_file.read().split('\n')[1:])
        readers.append(reader)
readers
```

```python
len(readers)
```

```python
col_count = len(readers)
row_count = len(readers)
reader_protein_sequence_similarity = [[0 for x in range(col_count)] for y in range(row_count)]
```

```python
from skbio import Protein
from skbio.alignment import local_pairwise_align_protein
for y in range(row_count):
    for x in range(col_count):
        query_seq = Protein(readers[x])
        target_seq = Protein(readers[y])
        try:
            alignment, score, start_end_positions = local_pairwise_align_protein(query_seq, target_seq, 11, 1)
        except:
            print('When some error occurs, x, y = {}, {}'.format(x, y))
        reader_protein_sequence_similarity[x][y] = score
```

```python
np.asarray(reader_protein_sequence_similarity)
```

```python
np.savetxt('Similarity_reader_protein_sequence.csv', np.asarray(reader_protein_sequence_similarity), delimiter='\t')
```

## Histone mark structural similarity (problematic)

```python
def pairwise2matrix(input_file, output_file): # convert 张若天's pairwise similarity tables to matrix form and store into a csv file
    import pandas as pd
    df = pd.read_table(input_file, sep=',', header=None)
    import math
    count = int(math.sqrt(len(df)))
    row_count = count
    col_count = count
    similarity_matrix = [[0 for x in range(col_count)] for y in range(row_count)]
    for y in range(row_count):
        for x in range(col_count):
            try:
                similarity_matrix[y][x] = float(df[4][df[0] == y][df[2] == x])
            except:
                print('When some error occurs, y, x = {}, {}'.format(y, x))
    import numpy as np
    np.savetxt(output_file, np.asarray(similarity_matrix), delimiter='\t')
    
```
