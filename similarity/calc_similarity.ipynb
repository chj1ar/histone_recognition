```python
import pandas as pd
from tqdm import tqdm
```

```python
uniquified_readers = pd.read_csv('uniquified_readers.csv', header=None)
```

```python
uniquified_histone_marks = pd.read_csv('uniquified_histone_marks.csv', header=None)
```

```python
uniquified_histone_marks_list = [x[0] for x in uniquified_histone_marks.values.tolist()]
```

```python
uniquified_histone_marks_list
```

```python
'''
id
Name	UniProtID	EnsemblID	Species	AA	Type	Substrate	Histone	Mark position	Flanking sequence	SMILES
'''
data_list[0]
```

```python
def assert_uniquified_histone_marks():
    for histone_mark in uniquified_histone_marks_list:
        template = None
        for row in data_list:
            if row[7] == histone_mark:
                template = row
                break
        if template is None:
            print('[E] %s not found'%histone_mark)
            continue
        for row in data_list:
            if row[7] == histone_mark:
                if row[-1]!= template[-1] or row[-2] != template[-2]:
                    print('[E] ', template, row)
                    print('[E] end')
```

```python
assert_uniquified_histone_marks()
```

```python
from skbio import Protein
from skbio.alignment import local_pairwise_align_protein

def seq_score(row_a, row_b ):
    s1 = Protein(row_a[-2])
    s2 = Protein(row_b[-2])
    alignment,score,start_end_positions = local_pairwise_align_protein(s1,s2,11,1)
    return score

from rdkit import Chem,DataStructs

def struct_score(row_a, row_b ):
    mol1 = Chem.MolFromSmiles(row_a[-1])
    mol2 = Chem.MolFromSmiles(row_b[-1])
    fp1 = Chem.RDKFingerprint(mol1)
    fp2 = Chem.RDKFingerprint(mol2)
    return DataStructs.TanimotoSimilarity(fp1,fp2)
```

```python
'i, mark_i, j, mark_j, seq_score, struct_score'
result = []
```

```python
for i in tqdm(range(len(uniquified_histone_marks_list))):
    mark_i = uniquified_histone_marks_list[i]
    template_i = None
    for row in data_list:
        if row[7] == mark_i:
            template_i = row
            break
    for j in range(len(uniquified_histone_marks_list)):
        mark_j = uniquified_histone_marks_list[j]
        template_j = None
        for row in data_list:
            if row[7] == mark_j:
                template_j = row
                break
        if template_i is None or template_j is None:
            result.append([i,mark_i,j,mark_j, None, None])
        else:
            result.append([i,mark_i,j,mark_j, seq_score(template_i, template_j), struct_score(template_i, template_j)])
        
```

```python
result
```

```python
# save
import csv
csvfile = open('histone_marks_seq_struct_similarity_result.csv', 'w') 
writer = csv.writer(csvfile)
writer.writerows(result)
csvfile.close()
```

```python
def uniprotid_eq(s1, s2):
    if ';' in s1:
        s1 = s1.split(';')
    else:
        s1 = [s1]
    if ';' in s2:
        s2 = s2.split(';')
    else:
        s2 = [s2]
    for x in s1:
        for y in s2:
            if x == y:
                return True
    return False
```
