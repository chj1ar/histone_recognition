```python
from skbio import Protein
from skbio.alignment import local_pairwise_align_protein

def seq_score(s):
    s1, s2 = s
    alignment,score,start_end_positions = local_pairwise_align_protein(s1,s2,gap_open_penalty=11,gap_extend_penalty=1)
    return score
```

```python
import skbio.sequence
# pr = skbio.sequence.Protein.read('./reader_sequences/Q9UBU8.fasta')
```

```python
prs = []
for i in range(len(uniquified_readers_list)):
    prs.append(skbio.sequence.Protein.read('./reader_sequences/%s.fasta'%uniquified_readers_list[i]))
assert(len(prs) == len(uniquified_readers_list))
```

```python
from multiprocessing import Pool
```

```python
prs[0]
```

```python
result = []
for i in tqdm(range(len(uniquified_readers_list))):
    r_this = []
    with Pool(15) as p: 
        r_this = p.map(seq_score, [(prs[i], prs[j]) for j in range(len(uniquified_readers_list))])
    for j in range(len(uniquified_readers_list)):
        result.append([i, uniquified_readers_list[i], j ,uniquified_readers_list[j], r_this[j]])
```

```python
result
```

```python
import csv
csvfile = open('reader_sequence_similarity_result_multiprocess.csv', 'w') 
writer = csv.writer(csvfile)
writer.writerows(result)
csvfile.close()
```
